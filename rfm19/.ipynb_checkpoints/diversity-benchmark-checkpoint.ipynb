```python
import csv, time
import numpy as np
import multihash as mh
import hashlib as hl
import matplotlib.pyplot as plt
import psycopg2
from binary_trie import Trie, bytes_to_bitstring, int_to_bitstring
```

```python
def multihash_to_bitstring(peer_id: str) -> str:  
    return bytes_to_bitstring(hl.sha256(mh.from_b58_string(peer_id)).digest())

def get_bucket_id(key0: str, key1: str) -> int:
    for i in range(len(key0)):
        if key0[i]!=key1[i]:
            return i
    return len(key0)
```

```python
with open("data/all-peerids.csv", 'r') as file:
    nebula_peerids = {line[0]:line[1] for line in csv.reader(file)}
```

```python
def compute_diversity(crawls_data):
    bucket_diversity_data = {i:{} for i in range(257)} # map{bucketID -> map{peerID -> occurence}}
    for line in crawls_data:
        peer_kad_id = multihash_to_bitstring(line[2])
        for neighbor_nebula in line[3:]:
            # compute bucket id
            neighbor = nebula_peerids[neighbor_nebula]
            neighbor_kad_id = multihash_to_bitstring(neighbor)
            bid = get_bucket_id(peer_kad_id, neighbor_kad_id)

            if neighbor not in bucket_diversity_data[bid]:
                bucket_diversity_data[bid][neighbor] = 1
            else:
                bucket_diversity_data[bid][neighbor] += 1
    return bucket_diversity_data
```

```python
def sort_diversity_data(diversity_data):
    sorted_diversity_data = {}
    for k in diversity_data:
        sorted_diversity_data[k] = [(diversity_data[k][peerid], peerid) for peerid in diversity_data[k]]
        sorted_diversity_data[k].sort(reverse=True)
    return sorted_diversity_data
```

```python
def bucket_diversity_to_file(diversity_data, week):
    filename = "data/bucket-diversity-week-"+str(week)
    
    lines = []
    for bucket in diversity_data:
        line = str(bucket)+":"
        for (occ, peerid) in diversity_data[bucket]:
            line+=str(occ)+","+peerid+";"
        lines.append(line[:-1]+'\n')
    
    f = open(filename, 'w')
    f.writelines(lines)
    f.close()
```

```python
def bucket_diversity_from_file(week):
    filename = "data/bucket-diversity-week-"+str(week)

    f = open(filename, 'r')
    lines = f.readlines()
    f.close()
        
    diversity_data = {}
    for line in lines:
        start_pos = line.find(':')
        bucket = int(line[:start_pos])
        splitted = line[start_pos+1:-1].split(';')
        values=[]
        for el in splitted:
            elements = el.split(',')
            if len(elements)>1:
                values.append((int(elements[0]),elements[1]))
        diversity_data[bucket]=values
    return diversity_data
```

```python
filename = "data/nebula-peers-1week-early.csv"
```

```python
with open(filename, 'r') as file:
    all_crawls = [line for line in csv.reader(file)]
```

```python
crawl_weeks = range(1,12)
for week in crawl_weeks:
    start_time = time.time()
    filename = "data/nebula-peers-week-"+str(week)+".csv"
    with open(filename, 'r') as file:
        crawls_data = [line for line in csv.reader(file)]
    diversity_data = compute_diversity(crawls_data)
    sorted_diversity = sort_diversity_data(diversity_data)
    bucket_diversity_to_file(sorted_diversity, week)
    print("Week",week,"done in",(time.time()-start_time)/60,"min")
```

```python
peers_per_week = []
distinct_peers_per_week = []
for week in crawl_weeks:
    start_time = time.time()
    filename = "data/nebula-peers-week-"+str(week)+".csv"
    with open(filename, 'r') as file:
        crawls_data = [line for line in csv.reader(file)]
    peers_per_week.append(len(crawls_data))
    
    distinct_peers = set()
    for line in crawls_data:
        distinct_peers.add(line[1])
    
    distinct_peers_per_week.append(len(distinct_peers))
```

```python
def bucket_cdf(bucket_occ):
    cdf = [0]
    count = 0
    curr = 1
    for v in reversed(bucket_occ):
        while v > curr:
            curr += 1
            cdf.append(count)
        
        count += v
    
    cdf = [v/count for v in cdf]
    return cdf
```

```python
steps = [10, 20, 50, 100, 500, 1000, 5000, 10000, 20000, 40000, 560000]
```

```python
def bucket_stats(bucket):
        
    
    index = len(steps)
    counts = [0 for _ in range(len(steps)+1)]
    c=0
    total = 0
    for (occ, _) in bucket:
        while index>0 and occ < steps[index-1]:
            index -= 1
        if index == len(steps):
            print(occ)
        counts[index] += occ
        total += occ
                
    return [c/total for c in counts]
```
